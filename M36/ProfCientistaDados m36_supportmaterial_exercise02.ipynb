## Criando variveis

**Variveis temporais**:  possvel que haja alguma regra de liberao de crditos de acordo com o dia do ms, o dia da semana, fim de semana (sim ou no), ou at alteraes temporais da regra. Vamos criar as seguintes variveis para testar estas hipteses:

- Dia da semana
- Dummy para fim de semana
- Dia do ms
- Ms (j na base)
- Trimestre
- Ano (j na base)

**Valor da nota**: Vamos criar tambm algumas variveis para resumir o valor da nota, e posteriormente criar uma intuio sobre categorizaes de variveis contnuas.

- Crie categorizaes da varivel ```Valor NF```, por quants, em 5, 10, 20 e 50 categorias.

````python
df['Dia da semana'] = df['Data Registro'].dt.dayofweek
df['Dia da semana'] = df['Dia da semana'].astype('category')
df['Fim de semana'] = df['Data Registro'].dt.dayofweek >=5
df['Dia do ms'] = df['Data Registro'].dt.day
df['Dia do ms'] = df['Dia do ms'].astype('category')
df['Ms'] = df['Data Registro'].dt.month
df['Ms'] = df['Ms'].astype('category')
df['Trimestre'] = df['Data Registro'].dt.quarter
df['Trimestre'] = df['Trimestre'].astype('category')
df['Valor NF 5'] = pd.qcut(df['Valor NF'], 5)
df['Valor NF 10'] = pd.qcut(df['Valor NF'], 10)
df['Valor NF 20'] = pd.qcut(df['Valor NF'], 20)
df['Valor NF 50'] = pd.qcut(df['Valor NF'], 50)
df.info()
````

## Calcule o IV de todas essas variveis

Sugesto: Construa um metadados e tente utilizar uma funo como a que foi utilizada durante a aula. Voc vai conseguir fazer esse clculo para as variveis uma a uma, mas em casos reais em que vamos mais alm na criao de variveis, podemos ter centenas, ou at milhares de variveis, esses clculos *precisam* ser automticos nesses casos.

````python
def IV(variavel, resposta):
    tab = pd.crosstab(variavel, resposta, margins=True, margins_name='total')

    rtulo_evento = tab.columns[0]
    rtulo_nao_evento = tab.columns[1]

    tab['pct_evento'] = tab[rtulo_evento]/tab.loc['total',rtulo_evento]
    tab['pct_nao_evento'] = tab[rtulo_nao_evento]/tab.loc['total',rtulo_nao_evento]
    tab['woe'] = np.log(tab.pct_evento/tab.pct_nao_evento)
    tab['iv_parcial'] = (tab.pct_evento - tab.pct_nao_evento)*tab.woe
    return tab['iv_parcial'].sum()

````

````python
variaveis = ['Dia da semana', 'Fim de semana', 'Dia do ms', 'Ms', 'Trimestre', 'Valor NF 5', 'Valor NF 10', 'Valor NF 20', 'Valor NF 50']
df_iv = pd.DataFrame()
for var in variaveis:
    df_iv.loc[var, 'IV'] = IV(df[var], df.evento)  

def colorir(val):
    cor = 'white'
    if val <= 0.02:
        cor = 'gray'
    elif val <= 0.1:
        cor = 'yellow'
    elif val <= 0.3:
        cor = 'cyan'
    elif val <= 0.5:
        cor = 'green'
    elif val > 0.5:
        cor = 'orange'
    return f'background-color: {cor}'

df_iv.style.applymap(colorir)
````

## Avalie o WOE e o IV no tempo

Avalie o WOE e o IV no tempo da varivel ```categoria```.

````python
df_tempo = df.copy()
df_tempo['Data Registro'] = df_tempo['Data Registro'].dt.to_period('M')
df_tempo = df_tempo.groupby(['Data Registro', 'categoria', 'evento'])['evento'].count()
df_tempo = df_tempo.to_frame()
df_tempo
````

````python
tab = df_tempo.unstack()['evento']
tab['N'] = tab.sum(axis=1)

tab['tx_evento'] = tab[True]/tab.N
tab.reset_index()
tab.head()
````

````python
rotulo_ev, rotulo_nev = (True, False)

tab['pct_evento']     = tab[rotulo_ev]/tab[rotulo_ev].groupby(level=0).sum()
tab['pct_nao_evento'] = tab[rotulo_nev]/tab[rotulo_nev].groupby(level=0).sum()
tab['WOE'] = np.log(tab.pct_evento/tab.pct_nao_evento)
tab
````

````python

fig, ax = plt.subplots(figsize=(12, 6))
variavel='categoria'

# Comando principal do grfico
sns.pointplot(data=tab.reset_index(), y = 'WOE', x = 'Data Registro', hue=variavel, ax=ax);

# ax.legend(bbox_to_anchor=(.6, -.1), ncol=2)

# Rotacionar os rtulos do eixo X
labs = tab.index.get_level_values(level=0).unique()
# labs2 = [x.to_period('M') for x in labs]

ax.set_xticklabels(labs,rotation=30, fontsize='small')
ax.legend(bbox_to_anchor=(.65, -.2), ncol=2)

#limites do grfico - mostrando 1 e -1 para ter referncia
ymax = max(1.2, max(tab.WOE))
ymin = min(-1.2, min(tab.WOE))
ax.set_ylim([ymin, ymax])

# Linhas de referncia
ax.axhline(1, linestyle=':'  , color='red', alpha=.5)
ax.axhline(0, linestyle='--' , color='grey', alpha=.5)
ax.axhline(-1, linestyle=':' , color='red', alpha=.5)

# Ttulos
ax.set_title(f'WOE ao longo do tempo para varivel {variavel.capitalize()}')
ax.set_xlabel('Tempo', loc='right');
````

````python
iv_categoria = ((tab.pct_evento - tab.pct_nao_evento)*tab.WOE).groupby(level=0).sum();
iv_categoria = iv_categoria[iv_categoria>0]
iv_categoria.index = iv_categoria.index.to_timestamp()
iv_categoria.plot.bar()
````

````python
def plot_iv(IV):
    fig, ax = plt.subplots(figsize=(12, 6))
    x = IV.index
    a = [.02]*len(IV.index)
    b = [.1]*len(IV.index)
    c = [.3]*len(IV.index)
    d = [.5]*len(IV.index)
    e = [.6]*len(IV.index)

    ax.fill_between(IV.index, a, color='grey', alpha=.2)
    ax.fill_between(IV.index, a, b, color='orange', alpha=.1)
    ax.fill_between(IV.index, b, c, color='green', alpha=.1)
    ax.fill_between(IV.index, c, d, color='blue', alpha=.1)
    ax.fill_between(IV.index, d, e, color='purple', alpha=.1)
    ax.plot(IV)

    labs = IV.index.get_level_values(level=0).unique()
    labs2 = [x.to_period('M') for x in labs]

    ax.set_xticks(labs)
    ax.set_xticklabels(labs2,rotation=30, fontsize='small')

    ax.set_title("Information Value de Categoria ao longo das safras")
    ax.set_ylabel("Information Value")
    ax.set_xlabel("Safra")
    
plot_iv(iv_categoria)
````
