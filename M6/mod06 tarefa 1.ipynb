```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("sinasc_RO_2019.csv")
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
plt.rc('figure', figsize=(8, 24))
fig, axes = plt.subplots(5, 1)

sns.countplot(df, x='APGAR1', ax=axes[0])
sns.countplot(df, x='APGAR5', ax=axes[1])
sns.countplot(df, x='SEXO', ax=axes[2])
sns.countplot(df, x='GESTACAO', ax=axes[3])
sns.countplot(df, x="ESTCIVMAE", ax=axes[4])
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
plt.rc('figure', figsize=(8, 6))
fig, axes = plt.subplots()

sns.barplot(data=df, x="APGAR5", y="CONSULTAS")
```
